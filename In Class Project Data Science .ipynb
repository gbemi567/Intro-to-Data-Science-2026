```python
# TASK 1: VARIABLES & DATA TYPES

# Integer
student_id = 1024

# String
student_name = "Jordan Lee"

# Float
gpa = 3.67

# Boolean
is_enrolled = True

# String
major = "Computer Science"

# Integer
credits_completed = 48

# Float
tuition_balance = 1520.75

# Boolean
honor_roll = True


# Display variable types
print("Student ID Type:", type(student_id))
print("Student Name Type:", type(student_name))
print("GPA Type:", type(gpa))
print("Enrollment Status Type:", type(is_enrolled))
print("Major Type:", type(major))
print("Credits Type:", type(credits_completed))
print("Tuition Balance Type:", type(tuition_balance))
print("Honor Roll Type:", type(honor_roll))
```

```python
# -----------------------------
# TASK 2: OPERATORS
# -----------------------------

# Grade components
homework = 85
midterm = 78
final_exam = 92
project = 88

# Weighted average calculation
weighted_average = (homework * 0.20) + (midterm * 0.25) + (final_exam * 0.35) + (project * 0.20)

print("Weighted Average:", weighted_average)

# Group formation example
total_students = 150
group_size = 12

full_groups = total_students // group_size
remaining_students = total_students % group_size

print("Full Groups:", full_groups)
print("Remaining Students:", remaining_students)

# Scholarship formula
gpa_value = 3.5
extra_activities = 40

scholarship_score = (gpa_value ** 2) + (extra_activities / 10)

print("Scholarship Score:", scholarship_score)
```

```python
# -----------------------------
# TASK 3: PRINTING METHODS
# -----------------------------

# Concatenation
print("Student: " + student_name + " | ID: " + str(student_id))

# Comma-separated
print("Student:", student_name, "GPA:", gpa)

# f-string (recommended modern method)
print(f"Student {student_name} (ID: {student_id}) has a GPA of {gpa}")

# .format() method
print("Student {} (ID: {}) has a GPA of {}".format(student_name, student_id, gpa))

# Formatting to 2 decimal places
print(f"Weighted Average: {weighted_average:.2f}")
```

```python
# -----------------------------
# TASK 4: LISTS
# -----------------------------

student_scores = [78, 92, 65, 88, 95, 72, 84, 90, 67, 83]

# Access elements
print("First Score:", student_scores[0])
print("Last Score:", student_scores[-1])

# Slicing
print("Scores 3 to 6:", student_scores[2:6])
print("Every Other Score:", student_scores[::2])

# Modify list
student_scores.append(91)
student_scores.insert(3, 77)
student_scores.remove(65)

print("Updated Scores:", student_scores)
```

```python
# -----------------------------
# TASK 5: DICTIONARIES
# -----------------------------

student_record = {
    "id": student_id,
    "name": student_name,
    "major": major,
    "gpa": gpa,
    "courses": ["Math", "English", "Biology"],
    "grades": {
        "Math": 90,
        "English": 85,
        "Biology": 88
    }
}

# Access data
print("Student Name:", student_record["name"])
print("Courses:", student_record["courses"])
print("Math Grade:", student_record["grades"]["Math"])

# Modify dictionary
student_record["advisor"] = "Dr. Thompson"
student_record["gpa"] = 3.80
student_record["courses"].append("Physics")
student_record["grades"]["Physics"] = 93

# Safe retrieval
print("Scholarship Status:", student_record.get("scholarship", "Not Assigned"))
```

```python
# -----------------------------
# TASK 6: TUPLES
# -----------------------------

grade_scale = ('A', 'B', 'C', 'D', 'F')

student_info = (student_id, student_name, gpa)

# Tuple unpacking
sid, sname, sgpa = student_info

print("Unpacked ID:", sid)
print("Unpacked Name:", sname)
print("Unpacked GPA:", sgpa)
```
